```python
import matplotlib.pyplot as plt
import pandas as pd

perth = pd.read_csv('perth.csv')
```

```python
price = perth['price']/1e6 # price per million

plt.figure(figsize = (15, 6))
plt.hist(price, bins = 50, color = 'tab:blue', alpha = 0.5, edgecolor = 'black')

plt.ylabel('Count', fontsize=25)
plt.xlabel('House Price (Million AUD)', fontsize=25)
plt.title(f'Perth House Prices {round(price.skew(), 2)}', fontsize=30)

plt.ylim([0, 3500])
plt.xlim([0, 2.5])

median = price.median()
plt.plot([median, median], [0, 3500], color="black")
plt.annotate(f'Median: {round(median, 2)}', (median-0.2, 2500), backgroundcolor = 'white')

mean = price.mean()
plt.plot([mean, mean], [0, 3500], color="black")
plt.annotate(f'Mean: {round(mean, 2)}', (mean+0.05, 2500), backgroundcolor = 'white')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.tight_layout()
```

```python
x = np.linspace(1, 2 * np.pi, 100)

y1 = np.sin(x)
y2 = np.exp(x)
```

```python
plt.figure(figsize=(10, 5))

# subplot
rows = 1
cols = 2

plt.subplot(rows, cols, 1)  # We want to control first fig
plt.plot(x, y1)
plt.xlabel("label 1")

plt.subplot(rows, cols, 2)  # We want to control second fig
plt.plot(x, y2)
plt.yscale("log")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(x, y1)
axs[1].plot(x, y2)

axs[0].set_xlabel("set xlabel 0")
axs[1].set_yscale("log")
```
