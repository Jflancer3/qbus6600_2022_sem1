```python
import matplotlib.pyplot as plt
import pandas as pd

perth = pd.read_csv('perth.csv')
```

```python
price = perth['price']/1e6 # price per million

plt.figure(figsize = (15, 6))
plt.hist(price, bins = 50, color = 'tab:blue', alpha = 0.5, edgecolor = 'black')

plt.ylabel('Count', fontsize=25)
plt.xlabel('House Price (Million AUD)', fontsize=25)
plt.title(f'Perth House Prices {round(price.skew(), 2)}', fontsize=30)

plt.ylim([0, 3500])
plt.xlim([0, 2.5])

mean = price.mean()
plt.plot([mean, mean], [0, 5000], color="black")
plt.annotate(f'Mean: {round(mean, 2)}', (mean+0.02, 2000), backgroundcolor='white', fontsize=20)

median = price.median()
plt.plot([median, median], [0, 5000], color="red")
plt.annotate(f'Median: {round(median, 2)}', (median-0.3, 2000), backgroundcolor='white', fontsize=20)

plt.tight_layout()
```

```python
x = np.linspace(0, 3, 100)

y1 = np.sin(x)
y2 = np.exp(x)
```

```python
rows = 1
cols = 2

plt.figure(figsize=(10, 5))

plt.subplot(rows, cols, 1)  # Controlling 1st plot
plt.plot(x, y1)
plt.xlabel("x-label 1")

plt.subplot(rows, cols, 2)  # Controlling 2nd plot
plt.plot(x, y2)
plt.yscale("log")
```

```python
rows = 1
cols = 2

fig, axs = plt.subplots(rows, cols, figsize=(10, 5))

axs[0].plot(x, y1)
axs[1].plot(x, y2)

axs[0].set_xlabel("x-label 0")
axs[1].set_yscale("log")
```
